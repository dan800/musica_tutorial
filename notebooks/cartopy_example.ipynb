```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
```

```python
File =  "/nobackup/pygjc/WACCM_UK_tutorial/Pre_industrial.nc" #file name
PI = xr.open_dataset(File,decode_times=False) #open the file and decode time as false
```

```python
'''
Mercator projection plot
'''

#create figure and set size
Figure = plt.figure(figsize=(5,5))   
#1st plot, create axis, pick projection, pick longitude, pick latitude
ax1 = plt.subplot(projection=ccrs.Mercator(central_longitude=0.0))
ax1.coastlines(resolution='110m') #include coastlines
#plot data
PI.TS.plot(cmap = 'coolwarm', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('Surface temperature', fontsize = 15)

#define colourbar function
#note that the plot has to be data = dataset.plot(...) for this to work
def cbar(label = True, clabel = 'Surface temperature'):
        cbar = plt.colorbar(data)
        if (label == True):
            clabel = clabel
            cbar.set_label(clabel,size=15)
        cbar.ax.tick_params(labelsize=15)

# now do the same plot but with levels and a colourbar
Figure = plt.figure(figsize=(6,4))   

# define subplot with mercator projection
ax1 = plt.subplot(projection=ccrs.Mercator(central_longitude=0.0))

ax1.coastlines(resolution='110m') #include coastlines
data = PI.TS.plot(cmap = 'seismic', levels = 11, #include levels 
                  add_labels=False, add_colorbar = False, #get rid of labels, get rid of colourbar
                  transform=ccrs.PlateCarree()) #include transform
plt.title('Pre-industrial atmosphere', fontsize = 15) #add title
cbar(label = True, clabel = 'Surface temperature') #include user defined colourbar
```

```python
# Zoom in on Europe
Figure = plt.figure(figsize=(5,5))   
ax = plt.subplot(projection=ccrs.EuroPP()) # EuroPP() zooms in on Europe
PI.TS.plot(cmap = 'coolwarm', levels = 11, add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
ax.coastlines(resolution='50m')  #include coastlines, change resoltion to 50 m
plt.title('Surface temperature', fontsize = 15)
```

```python
'''
Orthographic projection plots - multiple subplots
'''

#create figure and set size
Figure = plt.figure(figsize=(8, 8))   
#1st plot, create axis, pick projection, pick longitude, pick latitude
ax1 = plt.subplot2grid((2,2), (0,0), projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=90.0))
ax1.coastlines(resolution='110m') #include coastlines
#plot data
PI.TS.plot(cmap = 'coolwarm', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('North Pole surface temperature', fontsize = 15)

#2nd plot, create axis, pick projection, pick longitude, pick latitude
ax2 = plt.subplot2grid((2,2), (0,1), projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=90.0))
ax2.coastlines(resolution='110m') #include coastlines
#plot data
PI.ICEFRAC.plot(cmap = 'Blues_r', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('North Pole sea ice fraction', fontsize = 15)

#3rd plot, create axis, pick projection, pick longitude, pick latitude
ax3 = plt.subplot2grid((2,2), (1,0), projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=-90.0))
ax3.coastlines(resolution='110m') #include coastlines
#plot data
PI.TS.plot(cmap = 'coolwarm', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('South Pole surface temperature', fontsize = 15)

#4th plot, create axis, pick projection, pick longitude, pick latitude
ax4 = plt.subplot2grid((2,2), (1,1), projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=-90.0))
ax4.coastlines(resolution='110m') #include coastlines
#plot data
PI.ICEFRAC.plot(cmap = 'Blues_r', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('South Pole sea ice fraction', fontsize = 15)
```

```python
'''
Robinson projection plots - multiple subplots
'''

#create figure and set size
Figure = plt.figure(figsize=(10, 4)) 

#1st plot, create axis, pick projection, pick longitude
ax1 = plt.subplot2grid((1,2), (0,0), projection=ccrs.Robinson(central_longitude=0.0))
ax1.coastlines(resolution='110m') #include coastlines
#plot data
PI.TS.plot(cmap = 'coolwarm', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('Surface temperature', fontsize = 15)

#2nd plot, create axis, pick projection, pick longitude
ax2 = plt.subplot2grid((1,2), (0,1), projection=ccrs.Robinson(central_longitude=0.0))
ax2.coastlines(resolution='110m') #include coastlines
#plot data
PI.ICEFRAC.plot(cmap = 'Blues_r', add_labels=False, add_colorbar = False, transform=ccrs.PlateCarree())
#set title of subplot
plt.title('Sea ice fraction', fontsize = 15)
```
